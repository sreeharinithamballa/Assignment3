```python
import requests

# Function to fetch data from the URL
def fetch_data(url, headers, querystring):
    response = requests.get(url, headers=headers, params=querystring)
    data = response.json()
    return data

# Example URL
url = "https://linkedin-api8.p.rapidapi.com/"
querystring = {"username":"username"}
headers = {
	"X-RapidAPI-Key": "f3734e52c3msh0947d4c25d28c44p121078jsn523a8ad16e83",
	"X-RapidAPI-Host": "linkedin-api8.p.rapidapi.com"
}

# Fetching data
data = fetch_data(url, headers, querystring)

# Parsing JSON data
name = f"{data['firstName']} {data['lastName']}"
headline = data['headline']
location = data['geo']['full']
skills = [skill['name'] for skill in data['skills']]

# Displaying the dashboard
print("LinkedIn Profile Dashboard")
print("--------------------------")
print("Name:", name)
print("Headline:", headline)
print("Location:", location)
print("Skills:", ', '.join(skills))
```
